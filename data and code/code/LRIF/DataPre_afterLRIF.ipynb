```python
f1 = open(LRIF_FILE, 'r')
lines = f1.readlines()
f1.close()
title_list = []
score_list= []
for i in range(round(len(lines)/3)):
    title_list_temp = lines[i*3].split('Feature list: ')[1][0:-2].split(',')
    title_list_temp = [title_list_temp[j][2:-1] for j in range(len(title_list_temp))]
    string = '['
    for j in title_list_temp:
        string += j
        string += '  '
    string = string[:-2] + ']'
    title_list.append(string)
    score = float(lines[i*3+1].split('   I-Score: ')[1])
    score_list.append(str(round(score, 6)))
```
